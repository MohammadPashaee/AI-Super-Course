```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/linear_regression_data.xlsx")
```

```python
df_in1 = df[['input1']]
df_in2 = df[['input2']]
df_in3 = df[['input3']]
df_in = df.drop(labels=["output"], axis=1)

df_out = df[['output']]
```

```python
plt.scatter(df_in1, df_out)
```

```python
plt.scatter(df_in2, df_out)
```

```python
plt.scatter(df_in3, df_out)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(df_in, df_out)
```

```python
print(model.coef_, model.intercept_)
```

```python
yhat = model.predict(df_in)
```

```python
plt.scatter(df_out, yhat)
```

```python
plt.scatter(df_in1, df_out)
plt.scatter(df_in1, yhat)
```

```python
plt.scatter(df_in2, df_out)
plt.scatter(df_in2, yhat)
```

```python
plt.scatter(df_in3, df_out)
plt.scatter(df_in3, yhat)
```

```python
model.score(df_in, df_out)
```
